```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

#Reading the dataset
dataset = pd.read_csv('best-selling video games of all time updated.csv')

#Selecting X axis as rank and Y axis as sales
X = dataset[['Rank']]
Y = dataset['Sales']   

#Training and testing the model
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=20)

#Creating the model
model = LinearRegression()
model.fit(X_train, Y_train)

#Arranging the ranks from 1 to 50
ranks = np.arange(1, 51).reshape(-1, 1)

#Making predictions on the ranks
salesp = model.predict(ranks)

#Evaluating the model
r2 = r2_score(Y_test, model.predict(X_test))

print(f'R-squared: {r2}')

#Plotting the regression line and graph
plt.scatter(X_test, Y_test, color='black', label='Random Data')
plt.plot(ranks, salesp, color='blue', linewidth=3, label='Regression Line')
plt.title('Linear Regression: Rank vs. Sales')
plt.xlabel('Rank')
plt.ylabel('Sales (In Billions)')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

#Reading the dataset
dataset = pd.read_csv('best-selling video games of all time updated.csv')

#Selecting X axis as years and Y axis as sales
X = dataset[['Release Year']]  
Y = dataset['Sales']    

#Training and testing the model
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, random_state=44)

#Creating the model
model = LinearRegression()
model.fit(X_train, Y_train)

#Making predictions on release year
salesp = model.predict(X_test)

#Evaluating the model
r2 = r2_score(Y_test, salesp)

print(f'R-squared: {r2}')

# Plot the regression line
plt.scatter(X_test, Y_test, color='black', label='Random Data')
plt.plot(X_test, salesp, color='blue', linewidth=3, label='Regression Line')
plt.title('Linear Regression: Release Year vs. Sales')
plt.xlabel('Release Year')
plt.ylabel('Sales (In Ten Millions)')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load the dataset
df = pd.read_csv('best-selling video games of all time updated.csv')

# Assume the columns are named 'Year' and 'Sales'
X = df[['Release Year']]  # Independent variable (Year)
y = df['Sales']    # Dependent variable (Sales)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=44)

# Create and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
salesp = model.predict(X_test)

# Evaluate the model
r2 = r2_score(y_test, salesp)

print(f'R-squared: {r2}')

# Plot the regression line
plt.scatter(X, y, color='black', label='All Data')
plt.plot(X_test, salesp, color='blue', linewidth=3, label='Regression Line')
plt.title('Linear Regression: Release Year vs. Sales')
plt.xlabel('Release Year')
plt.ylabel('Sales (In Billions)')
plt.legend()
plt.show()
```
